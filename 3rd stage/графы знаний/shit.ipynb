```python
triples = None
with open('triplets_train.txt', 'r', encoding='utf-8') as file:
  triples = eval(file.read())
print(triples[:10])
```

```python
df['train'] = False
df.loc[:235-1, 'train'] = True
```

```python
from ampligraph.evaluation import train_test_split_no_unseen
X_train, X_valid = train_test_split_no_unseen(np.array(triples), test_size=65)  # 235*0.28
```

```python
from ampligraph.latent_features import ScoringBasedEmbeddingModel
from ampligraph.latent_features.loss_functions import get as get_loss
from ampligraph.latent_features.regularizers import get as get_regularizer

model = ScoringBasedEmbeddingModel(k=7,
                                   eta=2,
                                   scoring_type='ComplEx',
                                   seed=0)

# Optimizer, loss and regularizer definition
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)
loss = get_loss('multiclass_nll')
regularizer = get_regularizer('LP', {'p': 3, 'lambda': 1e-5})

# Compilation of the model
model.compile(optimizer=optimizer, loss=loss, entity_relation_regularizer=regularizer)
```

```python
model.fit(X_train,
          batch_size=int(X_train.shape[0] / 50),
          epochs=300, # Number of training epochs
          #verbose=False
          )
```

```python
ranks = model.evaluate(X_valid,
                      use_filter={'train': X_train,
                                  'test': X_valid},
                      corrupt_side='s,o',
                      verbose=True)
```

```python
from ampligraph.evaluation import mr_score, mrr_score, hits_at_n_score

mr = mr_score(ranks)
mrr = mrr_score(ranks)

print("MRR: %.2f" % (mrr))
print("MR: %.2f" % (mr))

hits_10 = hits_at_n_score(ranks, n=10)
print("Hits@10: %.2f" % (hits_10))
hits_3 = hits_at_n_score(ranks, n=3)
print("Hits@3: %.2f" % (hits_3))
hits_1 = hits_at_n_score(ranks, n=1)
print("Hits@1: %.2f" % (hits_1))
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from scipy.special import expit
# Разделение на тренировочную и тестовую выборки
triples = np.array(triples)
train_triples, test_triples = train_test_split(triples, test_size=0.2, random_state=42)

# Преобразование данных в DataFrame
df = pd.DataFrame(train_triples, columns=['game', 'attribute', 'value'])

# Выбор игры для фильтрации
target_game = "Grand Theft Auto V"
print("Все атрибуты игры:", df[df['game'] == target_game])
```

```python
# Фильтрация данных (удаляем некоторые атрибуты)
filtered_data = df[
    (df['game'] != target_game) |
    ((df['game'] == target_game) & ~df['attribute'].isin(["has_metacritic_score", "has_rating"]))
]

# Конвертация отфильтрованных данных обратно в массив
df_filtered = np.array(filtered_data)

# Эмуляция модели
class SimpleModel:
    def __init__(self, embedding_dim, entities, relations):
        self.embedding_dim = embedding_dim
        self.entity_embeddings = {e: np.random.rand(embedding_dim) for e in entities}
        self.relation_embeddings = {r: np.random.rand(embedding_dim) for r in relations}

    def fit(self, data):
        print("Фиктивное обучение на данных:", data.shape)

    def predict(self, statements):
        scores = []
        for s, p, o in statements:
            s_emb = self.entity_embeddings[s]
            p_emb = self.relation_embeddings[p]
            o_emb = self.entity_embeddings.get(o, np.random.rand(self.embedding_dim))
            score = np.dot(s_emb + p_emb, o_emb)
            scores.append(score)
        return np.array(scores)

# Уникальные сущности и отношения
entities = set(triples[:, 0]).union(triples[:, 2])
relations = set(triples[:, 1])

# Инициализация модели
model = SimpleModel(embedding_dim=50, entities=entities, relations=relations)

# Обучение модели на отфильтрованных данных
model.fit(df_filtered)

# Задание утверждений для предсказания
statements = np.array([
    [target_game, 'has_rating', 3.0],
    [target_game, 'has_rating', 4.0],
    [target_game, 'has_rating', 5.0],
    [target_game, 'has_genre', 'Action'],
    [target_game, 'has_genre', 'Shooter'],
])

# Предсказание вероятностей
scores = model.predict(statements)
probs = expit(scores)

# Преобразование результатов в DataFrame
result_df = pd.DataFrame(list(zip([' '.join(map(str, x)) for x in statements],
                                  scores,
                                  probs)),
                         columns=['statement', 'score', 'prob']).sort_values("prob", ascending=False)

print(result_df)
```
